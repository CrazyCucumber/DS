```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Создание простого графика

A = [1, 2, 3, 4, 5, 6, 9, 1, 2, 1, 5, 3, 2]
plt.plot(A)
plt.show()
```

```python
x = np.arange(0, 20)
y = np.random.randint(0, 30, 20)
plt.plot(x, y)
plt.show()
```

```python
#Кастом линиий и фигуры

a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])
c = np.array([3,2,4,6,5])

#Делает фигуру больше
plt.figure(figsize=(15,10))

#Кастом линий
plt.plot(a, color = 'green', linestyle = '-.', label="Rent FY20")
plt.plot(b, linewidth=5., label="AVG Rent")
plt.plot(c, color = 'red', linestyle = '--', label="Rent FY19")

#Добавление легенды
plt.legend(loc="upper left", fontsize=15)

#Добавление титульника
plt.title("Rent per Month Summary", fontsize=20, fontweight="bold")

#Замена обозначений по x
plt.xticks(range(0,len(a)),('Jan','Feb','Mar','Apr','May'))

#Добавление названий к осям
plt.xlabel("Months")
plt.ylabel("Rent $")

plt.show()
```

```python
#График функции разброса. Здесь доступны все те же функции, что и в plot

a = np.arange(30)
b = np.random.randint(0,30,30)

plt.scatter(a, b)
plt.show()
```

```python
#Тот же самый график, только теперь поменяли размер и цвет точек

#Массив из рандомных чисел для размера точек
s = np.random.randint(0,100,30)

plt.scatter(a,b, c="green", s=s)
plt.show()
```

```python
#Построение диаграмм. Здесь доступны все те же функции, что и в plot

#Данные для диаграмм
job = ['sportsman','teacher','programmer']
salary = np.array([4350, 5000, 7000])

#Изменение фигуры
plt.figure(figsize=(10,10))

#Отрисовка диаграммы. В первом элементе должна быть последовательность
plt.bar(range(0, len(job)), salary)

#Изменение названия осей
plt.xlabel("job")
plt.ylabel("salary $")

#Изменение координат по х
plt.xticks(range(0,len(job)),('sportsman','teacher','programmer'))

#Добавление загаловка
plt.title("Bar of salary", fontsize=15, fontweight="bold")

plt.show()
```

```python
#Построение сразу трёх фигур

#Изменение фигуры
plt.figure(figsize=(10,5))

#Отрисовка диаграммы. В первом элементе должна быть последовательность
plt.subplot(131)
plt.bar(range(0, len(job)), salary)
plt.xticks(range(0,3),job)

#Отрисовка линии по вершинам диаграммы
plt.subplot(132)
plt.plot(salary, c = 'red')
plt.xticks(range(0,3),job)

#Отрисовка вершин диаграммы
plt.subplot(133)
plt.scatter(range(0, 3), salary, c = 'green')
plt.xticks(range(0,3),job)

plt.show()
```
