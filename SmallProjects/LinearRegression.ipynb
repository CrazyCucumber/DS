```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
#Считываем данные

data = pd.read_csv('D:\\DATA\\DATA_WEIGHT.csv')
data.head()
```

```python
#Построим график

sns.scatterplot(x = data.Height, y = data.Weight)
```

```python
#Посмотрим всю самую важную информацию

data.describe()
```

```python
data.info()
```

```python
#Начнем делать нашу модель

model = LinearRegression()
```

```python
X = pd.DataFrame(data.Height)
Y = pd.DataFrame(data.Weight)
```

```python
#Делаем модель

model.fit(X, Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#Построим график

plt.scatter(x = data.Height, y = data.Weight)
plt.plot(X, model.predict(X), color = 'red')
```

```python
model.score(X, Y)
```
