### 6) Добавим столбец "В составе системы:". Он берется из шапки

```python
data_in_sys = data2["Наименование изделия"]
```

```python
data_in_sys
```

```python
#добавил ИМТ, хотя его в таблице не было
in_sys_new = []

for c in data_in_sys:
    if 'АКБ' in c or 'БП-3Г' in c or 'Блок питания' in c:
        in_sys_new.append('Питание')
    elif 'ИТ ПД-2' in c or 'ИТ-1-07' in c or 'ИТ-БПТ' in c or 'драйвер фазы' in c or 'БП-200' in c or 'БТР' in c or 'плата МУ-М' in c or 'тележка не идет на ход' in c or 'вагон не идет на ход' in c or 'датчик скорости' in c or 'ИМТ' in c:
        in_sys_new.append('Тяговый привод')
    elif 'Плата МУЗ-02' in c or 'Пружина ИМТ' in c or 'Тормозной диск' in c or 'направляющая ИМТ' in c or 'Направляющая тормозных колодок' in c or 'БУД-24' in c:
        in_sys_new.append('Тормозная')
    elif 'ВБ-11' in c or 'не включается тележка' in c:
        in_sys_new.append('Высоковольтное оборудование')
    elif 'Контроллер водителя' in c or 'КВ-8' in c or 'КВ-4' in c or 'Реверс' in c:
        in_sys_new.append('Управление')
    elif 'Датчик температуры' in c:
        in_sys_new.append('Климат')
    elif 'Дверь пассажирская' in c:
        in_sys_new.append('Доступ в салон')
    elif 'жесткий диск' in c or 'HDD' in c:
        in_sys_new.append('Видеорегистрация')
    elif 'редуктор' in c:
        in_sys_new.append('Трансмиссия')
    elif 'тэд пэмз' in c or 'Асинхронный двигатель' in c or 'тяговый двигатель' in c:
        in_sys_new.append('двигатель')
    else:
        in_sys_new.append('')
```

```python
print('Сделаем проверку:')
for i in range(len(data_in_sys)):
    print(data_in_sys[i], '-', in_sys_new[i], '\n')
```

```python
data2["В составе системы:"] = in_sys_new
```
