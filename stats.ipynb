```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Сделаем выборку из нормального распределения

norm_rv = sts.norm(loc=2, scale=0.5)
norm_rv.rvs(size=10)
```

```python
#Вернет значение функции распределения в переданной точке
print(norm_rv.cdf(4))
#Вернет значение функции плотности распределения в переданной точке
print(norm_rv.pdf(3))
```

```python
#Построим график нормального распределения

x=np.linspace(0,4,100)
cdf=norm_rv.cdf(x)
# функция может принимать и вектор (x)
plt.plot(x,cdf)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
#Построим график плотности нормального распределения

x=np.linspace(0,4,100)
pdf=norm_rv.pdf(x)
plt.plot(x,pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
